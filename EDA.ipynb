```python
import pandas as pd
import numpy as np
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)
```

```python
df = load_dataset("Car_data.csv")
df.head(5)
```

```python
df["name"] = df.name.apply(lambda x : ' '.join(x.split(' ')[:1]))
df['name']
```

```python
df.name.value_counts()
```

```python
sns.countplot(data=df,x="name",palette="CMRmap")
plt.xticks(rotation=90)
plt.xlabel("Name",fontsize=10,color="black")
plt.ylabel("Number",fontsize=10,color="black")
plt.title("NAME COUNT",color="black")
plt.show()
```

```python
labels = df["name"][:30].value_counts().index
sizes = df["name"][:30].value_counts()

plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',shadow=True, startangle=45)
plt.title('name',color = 'red',fontsize = 15)
plt.show()
```

```python
# from the above pie chart we can conform that the most common car is maruti
```

```python
df.year.value_counts()
```

```python
sns.countplot(data=df,x="year",palette="icefire")
plt.xticks(rotation=90)
plt.title("YEAR COUNT",color="DARKBLUE")
plt.show()
```

```python
labels = df["year"][:40].value_counts().index
sizes = df["year"][:40].value_counts()
plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, rotatelabels=False, autopct='%1.1f%%',shadow=True, startangle=45)
plt.title('Year',color = 'red',fontsize = 15)
plt.show()
```
